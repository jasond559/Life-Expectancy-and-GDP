# Import necessary libraries for analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

# Import data into dataframe

```python
df = pd.read_csv('all_data.csv')
df.head()
```

# Find any null values

```python
df.isnull().count()
```

# Find all countries in data

```python
df['Country'].unique()
```

# Update life expectancy column name

```python
df.rename(columns={'Life expectancy at birth (years)': 'life_expectancy'}, inplace=True)
```

# Graph life expectancy for all 6 countries

```python
sns.lineplot(y='life_expectancy', x='Year', data=df, hue='Country')
```

```python
g = sns.FacetGrid(df, col='Country', sharey=False, col_wrap=3, hue='Country')
g.map_dataframe(sns.lineplot, x='Year', y='life_expectancy')
g.set_titles(col_template='{col_name}');
```

# Graph GDP for all 6 countries

```python
sns.lineplot(x='Year', y='GDP', data=df, hue='Country')
```

```python
g = sns.FacetGrid(df, col='Country', sharey=False, col_wrap=3, hue='Country')
g.map_dataframe(sns.lineplot, x='Year', y='GDP')
g.set_titles(col_template='{col_name}')
g.set_ylabels('GDP (trillions of USD)');
```

# Scatter showing relationship with LE and GDP

```python
g = sns.FacetGrid(df, col='Country', sharex=False, sharey=False, col_wrap=3, hue='Country')
g.map_dataframe(sns.scatterplot, x='GDP', y='life_expectancy')
g.set_titles(col_template='{col_name} Life Expectancy vs GDP');
```

# Averages for Life Expectancy and GDP Among Countries

```python
sns.barplot(y='Country', x='life_expectancy', data=df)
```

```python
sns.barplot(y='Country', x='GDP', data=df)
```
